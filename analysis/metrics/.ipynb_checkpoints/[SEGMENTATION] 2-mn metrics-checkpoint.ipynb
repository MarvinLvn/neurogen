```python
from pathlib import Path
import pandas as pd
import numpy as np
results_folder = Path('/home/engaclew/neurogen/results/pyannote_metrics')

# Load f-score
fscore_lena = pd.read_csv(results_folder / 'its_eaf_an1' / 'fscore.csv')[['recording_id', 'group_id', 'retrieved','relevant','relevant retrieved','fscore']]
fscore_vtc = pd.read_csv(results_folder / 'vtc_eaf_an1' / 'fscore.csv')[['recording_id', 'group_id', 'retrieved','relevant','relevant retrieved','fscore']]

# Order
desired_order = ['low_risk', 'angelman_syndrome', 'fragile_x_syndrome', 'down_syndrome', 'autism_sibling']
fscore_lena['group_id'] = pd.Categorical(fscore_lena['group_id'], categories=desired_order, ordered=True)
fscore_vtc['group_id'] = pd.Categorical(fscore_vtc['group_id'], categories=desired_order, ordered=True)
fscore_lena = fscore_lena.sort_values('group_id')
fscore_vtc = fscore_vtc.sort_values('group_id')
print('Computing on %d chunks' % len(fscore_lena))
# Group data
def group_fscore(fscore_data):
    aggregator = np.mean
    fscore_data['precision'] = fscore_data['relevant retrieved'] / fscore_data['retrieved']
    fscore_data['recall'] = fscore_data['relevant retrieved'] / fscore_data['relevant']
    fscore_data = fscore_data.groupby('group_id').agg({
        'precision': [aggregator, 'std', 'count'],
        'recall': [aggregator, 'std'],
        'fscore': [aggregator, 'std'],
    }).reset_index()
    columns = ['group_id', 'precision_mean', 'precision_std', 'count', 'recall_mean', 'recall_std', 'fscore_mean', 'fscore_std']
    fscore_data.columns = columns
    for col in columns:
        if col not in ['group_id', 'count']:
            fscore_data[col] = 100*fscore_data[col]
    return fscore_data

fscore_lena = group_fscore(fscore_lena)
fscore_vtc = group_fscore(fscore_vtc)
```

```python
# SEGMENTATION SCORES
import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns

matplotlib.rcParams.update({'font.size': 16, 'legend.handleheight': 1, 'hatch.linewidth': 1.0,
                          'lines.markersize': 10, 'lines.linewidth': 2, 'errorbar.capsize': 5,
                           'xtick.labelsize': 13, 'ytick.labelsize': 13})

cm = 1/2.54


lena_color = '#6664b8'
aclew_color = '#E69F5C' 
metric_name = ['Precision (%)', 'Recall (%)', 'F-score (%)']
def plot_corpus_metrics(fscore_lena, fscore_vtc):
    metrics = ['precision', 'recall', 'fscore']
    fig_width = 6 * len(metrics)
    fig, axes = plt.subplots(1, len(metrics), figsize=(fig_width*cm, 12*cm))
    width = 0.35
    
    # Get means and stds for each metric at corpus level
    lena_means = [fscore_lena[f'{m}_mean'].mean() for m in metrics]
    lena_stds = [fscore_lena[f'{m}_std'].mean() for m in metrics]
    vtc_means = [fscore_vtc[f'{m}_mean'].mean() for m in metrics]
    vtc_stds = [fscore_vtc[f'{m}_std'].mean() for m in metrics]
    
    # Create each subplot
    for i, (metric, ax) in enumerate(zip(metrics, axes)):
        # Plot bars with error bars
        ax.bar(-width/2, 
               lena_means[i], 
               width, 
               yerr=lena_stds[i],
               label='LENA', 
               color=lena_color, 
               edgecolor='black', 
               zorder=2,
               error_kw=dict(ecolor='black', elinewidth=1.5, capsize=5, capthick=1))
               
        ax.bar(width/2, 
               vtc_means[i], 
               width,
               yerr=vtc_stds[i],
               label='ACLEW', 
               color=aclew_color, 
               edgecolor='black',
               zorder=2,
               error_kw=dict(ecolor='black', elinewidth=1.5, capsize=5, capthick=1))
        
        ax.set_xticks([])
        ax.set_ylabel(metric_name[i])
        ax.set_ylim(0, 100)
        ax.grid(True, linestyle='--', alpha=0.7, zorder=0)
        
        # Add value labels on top of bars
        ax.text(-width/2, lena_means[i]/2, f'{lena_means[i]:.0f}', 
                ha='center', va='center', color='white', weight='bold', zorder=3)
        ax.text(width/2, vtc_means[i]/2, 
                f'{vtc_means[i]:.0f}', 
                ha='center', va='center', color='white', weight='bold', zorder=3)
        
        ax.spines['top'].set_visible(False)
        ax.spines['right'].set_visible(False)
    handles = [plt.Rectangle((0,0),1,1, color=lena_color),  plt.Rectangle((0,0),1,1, color=aclew_color)]
    labels = ['LENA', 'ACLEW']
    fig.legend(handles, labels, ncol=2, bbox_to_anchor=(0.53, 1.03), loc='center')
    plt.tight_layout()
    plt.savefig('/home/engaclew/neurogen/figures/30mn/segmentation_metrics.png', dpi=300, bbox_inches='tight')
    return fig, axes

fig, axes = plot_corpus_metrics(fscore_lena, fscore_vtc)
plt.show()
```

```python
# Group-level metrics
lena_color = '#6664b8'
aclew_color = '#E69F5C' 

# Function to create grouped bar plots
def plot_metric(metric):
    fig, ax = plt.subplots(1, 1, figsize=(16*cm, 12*cm))
    metric_name = {
        'precision': 'Precision (%)',
        'recall': 'Recall (%)',
        'fscore': 'F-score (%)'
    }
    label_name = {
        'low_risk': 'Low\nrisk',
        'angelman_syndrome': 'Angelman',
        'fragile_x_syndrome': 'Fragile X',
        'down_syndrome': 'Down',
        'autism_sibling': 'Autism\nsibling'
    }
    x = range(len(fscore_lena))
    width = 0.35
    
    # Plot bars with error bars
    ax.bar([i - width/2 for i in x], 
           fscore_lena[f'{metric}_mean'], 
           width, 
           yerr=fscore_lena[f'{metric}_std'],
           label='LENA', 
           color=lena_color, 
           edgecolor='black', 
           zorder=2,
           error_kw=dict(ecolor='black', elinewidth=1.5, capsize=5, capthick=1))
           
    ax.bar([i + width/2 for i in x], 
           fscore_vtc[f'{metric}_mean'], 
           width,
           yerr=fscore_vtc[f'{metric}_std'],
           label='ACLEW', 
           color=aclew_color, 
           edgecolor='black',
           zorder=2,
           error_kw=dict(ecolor='black', elinewidth=1.5, capsize=5, capthick=1))
    
    ax.set_ylabel(metric_name[metric])
    ax.set_xticks(x)
    ax.set_xticklabels([label_name[group] for group in fscore_lena['group_id']], 
                       rotation=0)
    ax.legend(ncol=2, bbox_to_anchor=(0.5, 1.1), loc='center')
    ax.set_ylim(0, 100)
    ax.grid(True, linestyle='--', alpha=0.7, zorder=0)
    
    # Add value labels on top of bars
    y_shift = 7
    for i in x:
        ax.text(i - width/2, fscore_lena[f'{metric}_mean'].iloc[i]-y_shift, 
                f'{fscore_lena[f"{metric}_mean"].iloc[i]:.0f}', 
                ha='center', va='bottom', color='white', weight='bold', zorder=3)
        ax.text(i + width/2, fscore_vtc[f'{metric}_mean'].iloc[i]-y_shift, 
                f'{fscore_vtc[f"{metric}_mean"].iloc[i]:.0f}', 
                ha='center', va='bottom', color='white', weight='bold', zorder=3)
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

# Create subplots for each metric
plot_metric('precision')
plt.tight_layout()
plt.savefig('/home/engaclew/neurogen/figures/group_lena_vtc_precision.png', dpi=300)
plt.show()

plot_metric('recall')
plt.tight_layout()
plt.savefig('/home/engaclew/neurogen/figures/group_lena_vtc_recall.png', dpi=300)
plt.show()

plot_metric('fscore')
plt.tight_layout()
plt.savefig('/home/engaclew/neurogen/figures/group_lena_vtc_fscore.png', dpi=300)
plt.show()
```
